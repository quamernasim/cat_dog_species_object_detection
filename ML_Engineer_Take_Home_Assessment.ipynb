# MODEL IMPLEMENTATION:

```python
import os
import torch
import random
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
seed = 2023

torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)
```

```python
class Model(nn.Module):

  def __init__(self, num_species = 6):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])

    ### Initialize the required Layers

    #adaptive pooling applied to the feature map of the last convolutional layer
    self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
    #four parallel FC layers to predict objectness, cat/dog, species, and bounding box coordinates
    self.have_object = nn.Linear(512, 1)
    self.cat_or_dog = nn.Linear(512, 1)
    self.specie = nn.Linear(512, num_species)
    self.bbox = nn.Linear(512, 4)
    ### Initialize the required Layers

  def forward(self, input):
      out_backbone = self.backbone(input)
      ### Write Forward Calls for the Model
      out_avgpool = self.avgpool(out_backbone)
      out_avgpool = torch.flatten(out_avgpool, 1)
      
      #sigmoid for objectness and cat/dog used as binary classification
      have_object = torch.sigmoid(self.have_object(out_avgpool))
      cat_or_dog = torch.sigmoid(self.cat_or_dog(out_avgpool))
      specie = self.specie(out_avgpool)
      bbox = self.bbox(out_avgpool)

      return {
          "bbox": bbox,
          "object": have_object,
          "cat_or_dog": cat_or_dog,
          "specie": specie
      }
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
#encoding the labels, could be done using sklearn's LabelEncoder, but I am doing it manually, since it is a small dataset
animal_encoding = {
    'cat': 0,
    'dog': 1
}

specie_encoding = {
    0: {
        'abyssinian': 0,
        'birman': 1,
        'persian': 2
    },
    1: {
        'american_bulldog': 0,
        'american_pit_bull_terrier': 1,
        'basset_hound': 2,
        'beagle': 3,
        'chihuahua': 4,
        'pomeranian': 5
    }
}
```

```python
#check if the xml file is present for the image
def check_xml_file(path):
    return os.path.exists(path.replace("images", "labels") + ".xml")
```

```python
from PIL import Image
import torch

import torchvision.transforms as transforms


class CustomDataset():

  def __init__(self, dataset_path, images_list, train=False):

    #applying preprocessing on the images based on resnet architecture
    self.preprocess = transforms.Compose([
      transforms.Resize((224, 224)),
      transforms.ToTensor(),
    ])

    self.train = train
    self.augmentations = transforms.Compose([
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
        transforms.GaussianBlur(kernel_size=(3, 3), sigma=(0.1, 2.0)),
        transforms.RandomGrayscale(p=0.1)
    ])

    image_folder_path = os.path.join(dataset_path, "images")
    label_folder_path = os.path.join(dataset_path, "labels")

    self.image_paths = []
    self.label_paths = []

    for path in os.listdir(image_folder_path):
      name = path.split(os.sep)[-1].split(".")[0]
      if name in images_list:
        self.image_paths.append(pjoin(image_folder_path, path))
        #check if the xml file is present for the image
        if check_xml_file(pjoin(root, "labels", name)):
          self.label_paths.append(pjoin(label_folder_path, name+".xml"))
        else:
          self.label_paths.append(None)

  def __len__(self):
    return len(self.image_paths)

  def __getitem__(self, index):
    image = Image.open(self.image_paths[index]).convert("RGB")

    #get shape of the image
    h, w, c = np.array(image).shape
    image = self.preprocess(image)
    if self.train:

      image = self.augmentations(image)

    if self.label_paths[index]:
        label_xml = read_xml_file(self.label_paths[index])

        xmin = np.float32(label_xml["xmin"] / w)
        ymin = np.float32(label_xml["ymin"] / h)
        xmax = np.float32(label_xml["xmax"] / w)
        ymax = np.float32(label_xml["ymax"] / h)
        bbox = [xmin, ymin, xmax, ymax]

        is_object = 1

        cat_or_dog = animal_encoding[label_xml["cat_or_dog"].lower()]

        specie = specie_encoding[cat_or_dog][label_xml["specie"].lower()]

        label = {
            "bbox": bbox,
            "object": torch.tensor(is_object),
            "cat_or_dog": torch.tensor(cat_or_dog),
            "specie": torch.tensor(specie)
        }
    else:
        label = {
           #use -999 for the values which are not present so that we can filter them out later
            "bbox": [-999, -999, -999, -999],
            "object": torch.tensor(0),
            "cat_or_dog": torch.tensor(-999),
            "specie": torch.tensor(-999),
        }

    return image, label

def collate_fn(batch):
    """
    Collate function to be used when wrapping the dataset in a dataloader.
    """

    images = list()
    label = list()
    for b in batch:
        images.append(b[0])
        label.append(b[1])

    images = torch.stack(images, dim=0)

    return images, label
```

```python
def post_process_object(x, thresh = 0.5):
  # setting threshold to 0.5, can be changed
  x = (x>=thresh)*1
  return x

def post_process_cat_or_dog(x, thresh = 0.5):
   # setting threshold to 0.5, can be changed
  x = (x>=thresh)*1
  return x

def post_process_specie(x):
  #getting the class with highest probability
  x = torch.argmax(torch.softmax(x, dim=1), dim=1)
  return x

def post_process_bbox(x):
  return x
```

```python
# import our library
import torchmetrics

def train(epochs, batch_size = 32, lr = 0.001, model_weights = None):

  # Initialize Model and Optimizer
  model = Model()
  optimizer = torch.optim.Adam(model.parameters(), lr=lr)

  # Initialize Loss Functions
  have_object_loss = torch.nn.BCELoss()
  specie_loss = torch.nn.CrossEntropyLoss()
  cat_or_dog_loss = torch.nn.BCELoss()
  bbox_loss = torch.nn.L1Loss() # Not necessary you need to apply function to all coordinates together, You can have separete loss functions for all coordinates too

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list, train = True)
  training_loader = DataLoader(training_dataset, batch_size=batch_size, collate_fn=collate_fn, shuffle=True, num_workers=2)

  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
  model = model.to(device)
  
  # if model_weights is provided then resume training from that epoch
  if model_weights:
    checkpoint = torch.load(model_weights, map_location=device)
    model.load_state_dict(checkpoint["state_dict"])
    optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
    start_epoch = checkpoint["epoch"] + 1
    best_loss = checkpoint["best_loss"]
    best_epoch = checkpoint["best_epoch"]
  else:
    start_epoch = 0
    best_epoch = 0
    best_loss = 99999999

  def train_one_epoch(epoch_index, tb_writer = None):
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      
      for i, data in tqdm(enumerate(training_loader), total = len(training_loader)):
          # Every data instance is an input + label pair
          inputs, labels = data
          inputs = inputs.to(device)

          # Make predictions for this batch
          outputs = model(inputs)

          # Compute the loss and its gradients
          # compute loss for object head        
          labels_have_object = torch.tensor([i["object"] for i in labels]).unsqueeze(-1).float().to(device)
          loss_have_object = have_object_loss(outputs["object"], labels_have_object)
          
          # compute loss for cat_or_dog head, do this only for those images where object is present
          lbl_cat_or_dog = torch.tensor([[i["cat_or_dog"], idx] for idx, i in enumerate(labels) if i["object"]])
          lbl_cat_or_dog, idx = lbl_cat_or_dog[:, 0].to(device), lbl_cat_or_dog[:, 1]
          out_cat_or_dog = outputs['cat_or_dog'][idx]
          loss_cat_or_dog = cat_or_dog_loss(out_cat_or_dog.float().squeeze(), lbl_cat_or_dog.float())

          # compute loss for specie head, do this only for those images where object is present
          lbl_specie = torch.tensor([i["specie"] for i in labels if i["object"]]).to(device)
          out_specie = outputs['specie'][idx]
          loss_specie = specie_loss(out_specie, lbl_specie)
          
          # compute loss for bbox head, do this only for those images where object is present
          lbl_bbox = torch.tensor([i["bbox"] for i in labels if i["object"]]).to(device)
          out_bbox = outputs['bbox'][idx]
          loss_bbox = bbox_loss(out_bbox, lbl_bbox)
          
          # Consolidate all individual losses, can use weights here
          loss = loss_have_object + loss_cat_or_dog + loss_specie + loss_bbox

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / 10 # loss per batch
              running_loss = 0.
      return last_loss

  for i in range(start_epoch, epochs):

    epoch_loss = train_one_epoch(i)
    print(f' Epoch {i} Loss : {epoch_loss}')

    if epoch_loss < best_loss:
        #though this is not right, we should update the best_loss using validation
        #doing this becuase of less time
        best_loss = epoch_loss
        best_epoch = i
        print("Saving Model")
        torch.save({"state_dict": model.state_dict(), 
                    "best_loss": best_loss, 
                    "epoch": i,
                    "optimizer_state_dict": optimizer.state_dict(),
                    "best_epoch": i}, "model.pth")
    if i - best_epoch>=8:
      print('Early Stopping')
      break
    metrics = test(model, val_list, batch_size, device)
    print(metrics)

def test(model, val_list, batch_size, device):

  val_dataset = CustomDataset("/content/assessment_dataset", images_list=val_list)
  val_loader = DataLoader(val_dataset, batch_size=batch_size, collate_fn=collate_fn, shuffle=True, num_workers=2)

  metric_object = torchmetrics.Accuracy('binary')
  metric_cat_or_dog = torchmetrics.Accuracy('binary')
  metric_specie = torchmetrics.Accuracy('multiclass', num_classes = 6)
  metric_bbox = None

  out_object_classes = []
  out_cat_or_dog_classes = []
  out_specie_classes = []
  out_bbox_classes = []

  label_object_classes = []
  label_cat_or_dog_classes = []
  label_specie_classes = []
  label_bbox_classes = []


  for i, data in tqdm(enumerate(val_loader), total = len(val_loader)):
    inputs, labels = data

    # Make predictions for this batch
    outputs = model(inputs.to(device))
    
    # get labels for object head
    labels_have_object = torch.tensor([i["object"] for i in labels]).unsqueeze(-1).float()
    
    # get labels for cat_or_dog head for those images where object is present
    lbl_cat_or_dog = torch.tensor([[i["cat_or_dog"], idx] for idx, i in enumerate(labels) if i["object"]])
    lbl_cat_or_dog, idx = lbl_cat_or_dog[:, 0], lbl_cat_or_dog[:, 1]
    out_cat_or_dog = outputs['cat_or_dog'][idx]

    # get labels for specie head for those images where object is present
    lbl_specie = torch.tensor([i["specie"] for i in labels if i["object"]])
    out_specie = outputs['specie'][idx]

    # get labels for bbox head for those images where object is present
    lbl_bbox = torch.tensor([i["bbox"] for i in labels if i["object"]])
    out_bbox = outputs['bbox'][idx]

    #apply post processing to get the class and bbox
    object_class = post_process_object(outputs['object'])
    cat_or_dog_class = post_process_cat_or_dog(out_cat_or_dog)
    specie_class = post_process_specie(out_specie)
    bbox_class = post_process_bbox(out_bbox)

    out_object_classes.append(object_class)
    out_cat_or_dog_classes.append(cat_or_dog_class)
    out_specie_classes.append(specie_class)
    out_bbox_classes.append(bbox_class)

    label_object_classes.append(labels_have_object)
    label_cat_or_dog_classes.append(lbl_cat_or_dog)
    label_specie_classes.append(lbl_specie)
    label_bbox_classes.append(lbl_bbox)

  # Concatenate all the batch outputs to one array
  out_object_classes = torch.cat(out_object_classes, dim=0)
  out_cat_or_dog_classes = torch.cat(out_cat_or_dog_classes, dim=0)
  out_specie_classes = torch.cat(out_specie_classes, dim=0)
  out_bbox_classes = torch.cat(out_bbox_classes, dim=0)

  # Concatenate all the batch labels to one array
  label_object_classes = torch.cat(label_object_classes, dim=0)
  label_cat_or_dog_classes = torch.cat(label_cat_or_dog_classes, dim=0)
  label_specie_classes = torch.cat(label_specie_classes, dim=0)
  label_bbox_classes = torch.cat(label_bbox_classes, dim=0)

  score_object = metric_object(out_object_classes.detach().cpu(), label_object_classes)
  score_cat_or_dog = metric_cat_or_dog(out_cat_or_dog_classes.squeeze().detach().cpu(), label_cat_or_dog_classes)
  score_specie = metric_specie(out_specie_classes.squeeze().detach().cpu(), label_specie_classes)
  # score_bbox = metric_bbox()

  return score_object, score_cat_or_dog, score_specie#, score_bbox
```

```python
#get key based on value
def get_key(val, my_dict):
    for key, value in my_dict.items():
         if val == value:
             return key
```

```python
from PIL import ImageDraw

def visualize(model_weights, image_name, device = 'cpu', output_folder="output"):

  model = Model()
  model.load_state_dict(torch.load(model_weights)['state_dict'])
  model = model.to(device)

  try:
    image = Image.open(os.path.join("/content/assessment_dataset/images", image_name+".jpg"))
  except:
    image = Image.open(os.path.join("/content/assessment_dataset/images", image_name+".jpeg"))

  #apply same preprocessing
  preprocess = transforms.Compose([
      transforms.Resize((224, 224)),
      transforms.ToTensor()
    ])
  
  image = preprocess(image).to(device)
  output = model(image.unsqueeze(0))
  image = transforms.ToPILImage()(image.cpu())

  out_object = output["object"].item()
  object_class = post_process_object(out_object)
  draw = ImageDraw.Draw(image)

  if object_class:
    out_cat_or_dog = output["cat_or_dog"]
    out_cat_or_dog_class = post_process_cat_or_dog(out_cat_or_dog).item()

    out_specie = output["specie"]
    out_specie_class = post_process_specie(out_specie).item()

    out_bbox = output["bbox"]*224
    out_bbox = post_process_bbox(out_bbox)[0].cpu().detach().numpy()

    xmin, ymin, xmax, ymax = out_bbox

    #draw image
    os.makedirs(output_folder, exist_ok=True)
    draw.rectangle(((xmin, ymin), (xmax, ymax)), outline="red", width=3)
    draw.text((xmin, ymin-20), f"Class: {get_key(out_cat_or_dog_class, animal_encoding)}", fill="black")
    draw.text((xmin, ymin-10), f"Specie: {get_key(out_specie_class, specie_encoding[out_cat_or_dog_class])}", fill="black")
  else:
    print("No Object Found")
  image.save(os.path.join(output_folder, image_name+".jpg"))
  return {}
```
